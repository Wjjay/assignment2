```python
import pandas as pd

data = pd.read_csv("yearly_temperature_change_by_country.csv")
data.head()
```

```python
new_df = pd.melt(data,id_vars=['Area'],var_name='Year', value_name='Temperature Change')
```

```python
new_df = new_df.rename(column={"Area":"Country Name"})
new_df.head()
```

```python
new_df.to_csv()
```
